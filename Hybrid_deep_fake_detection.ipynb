```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import cv2
from tqdm import tqdm

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Define paths for dataset and model
dataset_path = '/content/drive/MyDrive/deepfake_images'
model_path = '/content/drive/MyDrive/deepfake_model.h5'

# Check if dataset exists
if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset directory not found at {dataset_path}")

print("Dataset contents:", os.listdir(dataset_path))

# Paths for real and fake image folders
real_images_path = os.path.join(dataset_path, 'real')
fake_images_path = os.path.join(dataset_path, 'fake')

# Split data into training and validation sets
from sklearn.model_selection import train_test_split

# Create temp directories for validation data if they don't exist
validation_path = os.path.join(dataset_path, 'validation')
if not os.path.exists(validation_path):
    os.makedirs(validation_path)
    os.makedirs(os.path.join(validation_path, 'real'))
    os.makedirs(os.path.join(validation_path, 'fake'))

# Setup data generators with different preprocessing for train and validation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # 20% for validation
)

# Only rescaling for validation
val_datagen = ImageDataGenerator(rescale=1./255)

# Prepare data generators
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=16,  # Smaller batch size for better generalization
    class_mode='binary',
    subset='training',  # Specify training subset
    shuffle=True
)

validation_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=16,
    class_mode='binary',
    subset='validation',  # Specify validation subset
    shuffle=False  # Don't shuffle validation data
)

print(f"Found {train_generator.samples} training images")
print(f"Found {validation_generator.samples} validation images")

# Check class indices to ensure proper labeling
print("Class indices:", train_generator.class_indices)

# Load the VGG19 model pre-trained on ImageNet
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the lower layers but unfreeze some higher layers for fine-tuning
for layer in base_model.layers[:-4]:  # Freeze all except last 4 layers
    layer.trainable = False

# Create a more robust model architecture
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1024, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # Binary output (Real or Fake)
])

# Use a lower learning rate for transfer learning
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

# Compile the model
model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(), tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

# Display model summary
model.summary()

# Set up callbacks for better training
checkpoint = ModelCheckpoint(
    model_path,
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    epochs=30,  # More epochs with early stopping
    callbacks=[checkpoint, early_stopping]
)

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/training_history.png')
plt.show()

# Load the best model
from tensorflow.keras.models import load_model
model = load_model(model_path)

# Function to extract frames from video
def extract_frames(video_path, output_folder, frames_per_second=1):
    """
    Extract frames from a video file.

    Args:
        video_path: Path to the video file
        output_folder: Folder to save the frames
        frames_per_second: Number of frames to extract per second

    Returns:
        List of paths to extracted frames
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidcap = cv2.VideoCapture(video_path)
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Extract frame every nth frame
    frame_interval = int(fps / frames_per_second)

    frame_count = 0
    extracted_count = 0
    frame_paths = []

    print(f"Extracting frames from video (total frames: {total_frames})")

    while True:
        success, frame = vidcap.read()
        if not success:
            break

        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{extracted_count:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_paths.append(frame_path)
            extracted_count += 1

        frame_count += 1

        # Show progress
        if frame_count % 100 == 0:
            print(f"Processed {frame_count}/{total_frames} frames")

    vidcap.release()
    print(f"Extracted {extracted_count} frames to {output_folder}")
    return frame_paths

# Function to predict if a frame is real or fake
def predict_frame(model, frame_path):
    """
    Predict if a frame is real or fake.

    Args:
        model: Trained model
        frame_path: Path to the frame

    Returns:
        Probability that the frame is fake (0-1)
    """
    # Load and preprocess the image
    img = tf.keras.preprocessing.image.load_img(frame_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Get model prediction
    pred = model.predict(img_array, verbose=0)[0][0]
    return float(pred)

# Function to analyze a video
def analyze_video(model, video_path, temp_folder="/content/temp_frames"):
    """
    Analyze a video for deepfake detection.

    Args:
        model: Trained model
        video_path: Path to the video
        temp_folder: Folder to save temporary frames

    Returns:
        Dictionary with analysis results
    """
    # Extract frames
    frame_paths = extract_frames(video_path, temp_folder)

    # Analyze frames
    frame_predictions = []
    print("\nAnalyzing frames...")
    for frame_path in tqdm(frame_paths):
        pred = predict_frame(model, frame_path)
        frame_predictions.append(pred)

    # Calculate statistics
    avg_score = np.mean(frame_predictions)

    # Use calibrated threshold (from ROC analysis on validation set)
    # This should be determined from validation data, but 0.6 is an example
    threshold = 0.6

    binary_preds = [1 if p > threshold else 0 for p in frame_predictions]
    fake_count = sum(binary_preds)
    total = len(binary_preds)
    fake_percentage = (fake_count / total) * 100

    # Get confidence level
    confidence = abs(avg_score - 0.5) * 2  # Scale to 0-1

    # Group frames by their prediction for vizualization
    bins = np.linspace(0, 1, 11)  # 10 bins
    hist, _ = np.histogram(frame_predictions, bins=bins)

    # Plot histogram of frame predictions
    plt.figure(figsize=(10, 6))
    plt.hist(frame_predictions, bins=bins, alpha=0.7)
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Distribution of Frame Predictions')
    plt.xlabel('Prediction Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.savefig('/content/drive/MyDrive/frame_prediction_distribution.png')

    return {
        'frame_predictions': frame_predictions,
        'fake_count': fake_count,
        'total_frames': total,
        'fake_percentage': fake_percentage,
        'avg_score': avg_score,
        'confidence': confidence,
        'is_fake': fake_percentage > 50
    }

# Example of how to use the analyze_video function:
# video_path = '/content/drive/MyDrive/test_video.mp4'
# results = analyze_video(model, video_path)

# Print results in a visually appealing way
def print_results(results):
    """Print the results of video analysis"""
    print("\n" + "="*50)
    print("🔍 DEEPFAKE DETECTION RESULTS 🔍")
    print("="*50)

    print(f"🧠 Detected {results['fake_count']}/{results['total_frames']} frames as potentially FAKE")
    print(f"📊 Fake percentage: {results['fake_percentage']:.2f}%")
    print(f"⚖️ Average prediction score: {results['avg_score']:.4f}")
    print(f"🎯 Confidence level: {results['confidence']:.2f}")

    if results['is_fake']:
        print("\n🔴 VERDICT: This video is classified as FAKE")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")
    else:
        print("\n🟢 VERDICT: This video is classified as REAL")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")

    print("="*50)
    print("📈 Check the saved distribution chart for more details")
    print("="*50)

# Example usage:
# video_path = '/content/drive/MyDrive/test_video.mp4'
# results = analyze_video(model, video_path)
# print_results(results)

# To analyze a specific video, uncomment and run:
"""
video_path = '/content/drive/MyDrive/test_video.mp4'  # Replace with your video path
results = analyze_video(model, video_path)
print_results(results)
"""

print("Model and analysis functions are ready. To analyze a video, call:")
print("results = analyze_video(model, '/path/to/your/video.mp4')")
print("print_results(results)")
```

```python
# Code for loading and using your deepfake detection model in future Colab sessions

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
from google.colab import drive, files
from tqdm import tqdm
import shutil
import time

# Mount Google Drive where your model is stored
drive.mount('/content/drive')

def load_deepfake_model(model_path=None):
    """
    Load the saved deepfake detection model

    Args:
        model_path: Path to the saved model (.h5 file)
                   If None, will prompt for file upload

    Returns:
        Loaded Keras model
    """
    if model_path is None or not os.path.exists(model_path):
        # Ask if model is in Google Drive or needs to be uploaded
        print("Select how to load your model:")
        print("1. Load from Google Drive")
        print("2. Upload model file")

        choice = input("Enter your choice (1 or 2): ")

        if choice == "1":
            # Search for model files in Google Drive
            print("\nSearching for model files in Google Drive...")
            drive_path = "/content/drive/MyDrive"
            model_files = [f for f in os.listdir(drive_path) if f.endswith('.h5')]

            if not model_files:
                print("❌ No .h5 model files found in Google Drive root.")
                print("📂 Please specify the exact path or upload a model file.")
                return None

            print("\nFound these model files in Google Drive:")
            for i, f in enumerate(model_files):
                print(f"{i+1}. {f}")

            idx = int(input("\nEnter the number of the model to load: ")) - 1
            if 0 <= idx < len(model_files):
                model_path = os.path.join(drive_path, model_files[idx])
            else:
                print("❌ Invalid selection.")
                return None
        else:
            # Upload model file
            print("\n📤 Please upload your .h5 model file...")
            uploaded = files.upload()

            if not uploaded:
                print("❌ No file was uploaded.")
                return None

            model_path = list(uploaded.keys())[0]

    # Load the model
    print(f"🔄 Loading model from {model_path}...")
    try:
        model = load_model(model_path)
        print("✅ Model loaded successfully!")
        return model
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return None

# Function to extract frames from video
def extract_frames(video_path, output_folder, frames_per_second=1):
    """
    Extract frames from a video file.

    Args:
        video_path: Path to the video file
        output_folder: Folder to save the frames
        frames_per_second: Number of frames to extract per second

    Returns:
        List of paths to extracted frames
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidcap = cv2.VideoCapture(video_path)
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Extract frame every nth frame
    frame_interval = int(fps / frames_per_second)

    frame_count = 0
    extracted_count = 0
    frame_paths = []

    print(f"Extracting frames from video (total frames: {total_frames})")

    while True:
        success, frame = vidcap.read()
        if not success:
            break

        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{extracted_count:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_paths.append(frame_path)
            extracted_count += 1

        frame_count += 1

        # Show progress
        if frame_count % 100 == 0:
            print(f"Processed {frame_count}/{total_frames} frames")

    vidcap.release()
    print(f"Extracted {extracted_count} frames to {output_folder}")
    return frame_paths

# Function to analyze frames in batches
def batch_analyze_frames(model, frame_paths, batch_size=16):
    """
    Process frames in batches to avoid memory issues with large videos

    Args:
        model: Trained deepfake detection model
        frame_paths: List of paths to video frames
        batch_size: Number of frames to process at once

    Returns:
        List of prediction scores for each frame
    """
    from tensorflow.keras.preprocessing import image

    all_predictions = []

    # Process frames in batches
    for i in tqdm(range(0, len(frame_paths), batch_size)):
        batch_paths = frame_paths[i:i+batch_size]
        batch_images = []

        # Prepare batch
        for frame_path in batch_paths:
            img = image.load_img(frame_path, target_size=(224, 224))
            img_array = image.img_to_array(img) / 255.0
            batch_images.append(img_array)

        # Convert to numpy array
        batch_data = np.array(batch_images)

        # Get predictions
        batch_preds = model.predict(batch_data, verbose=0)

        # Store predictions
        all_predictions.extend([float(p[0]) for p in batch_preds])

    return all_predictions

# Function to analyze a video
def analyze_video(model, video_path, temp_folder="/content/temp_frames", batch_size=16):
    """
    Analyze a video for deepfake detection.

    Args:
        model: Trained model
        video_path: Path to the video
        temp_folder: Folder to save temporary frames
        batch_size: Batch size for processing frames

    Returns:
        Dictionary with analysis results
    """
    # Extract frames
    frame_paths = extract_frames(video_path, temp_folder)

    if not frame_paths:
        raise ValueError("No frames could be extracted from the video")

    # Analyze frames in batches
    print("\nAnalyzing frames in batches...")
    frame_predictions = batch_analyze_frames(model, frame_paths, batch_size)

    # Calculate statistics
    avg_score = np.mean(frame_predictions)

    # Use 0.3 as the threshold, as requested
    threshold = 0.3

    binary_preds = [1 if p > threshold else 0 for p in frame_predictions]
    fake_count = sum(binary_preds)
    total = len(binary_preds)
    fake_percentage = (fake_count / total) * 100

    # Get confidence level
    confidence = abs(avg_score - 0.5) * 2  # Scale to 0-1

    # Group frames by their prediction for visualization
    bins = np.linspace(0, 1, 11)  # 10 bins
    hist, _ = np.histogram(frame_predictions, bins=bins)

    # Plot histogram of frame predictions
    plt.figure(figsize=(10, 6))
    plt.hist(frame_predictions, bins=bins, alpha=0.7)
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Distribution of Frame Predictions')
    plt.xlabel('Prediction Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()

    save_path = '/content/frame_prediction_distribution.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/frame_prediction_distribution.png'

    plt.savefig(save_path)
    plt.show()

    return {
        'frame_predictions': frame_predictions,
        'fake_count': fake_count,
        'total_frames': total,
        'fake_percentage': fake_percentage,
        'avg_score': avg_score,
        'confidence': confidence,
        'is_fake': fake_percentage > 50
    }

# Print results in a visually appealing way
def print_results(results):
    """Print the results of video analysis"""
    print("\n" + "="*50)
    print("🔍 DEEPFAKE DETECTION RESULTS 🔍")
    print("="*50)

    print(f"🧠 Detected {results['fake_count']}/{results['total_frames']} frames as FAKE")
    print(f"📊 Fake percentage: {results['fake_percentage']:.2f}%")
    print(f"⚖️ Average prediction score: {results['avg_score']:.4f}")
    print(f"🎯 Confidence level: {results['confidence']:.2f}")

    if results['is_fake']:
        print("\n🔴 VERDICT: This video is classified as FAKE")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")
    else:
        print("\n🟢 VERDICT: This video is classified as REAL")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")

    print("="*50)
    print("📈 Check the saved distribution chart for more details")
    print("="*50)

# Function to visualize sample frames with their predictions
def plot_sample_frames(frames_dir, predictions, num_samples=5):
    """
    Plot sample frames from the video with their prediction scores

    Args:
        frames_dir: Directory containing video frames
        predictions: List of prediction scores for each frame
        num_samples: Number of sample frames to show
    """
    import cv2
    import numpy as np

    # Get all frame files
    frame_files = sorted([f for f in os.listdir(frames_dir) if f.endswith('.jpg')])

    if len(frame_files) == 0:
        print("No frames found to display")
        return

    # Select sample frames (evenly distributed)
    if len(frame_files) <= num_samples:
        sample_indices = list(range(len(frame_files)))
    else:
        sample_indices = np.linspace(0, len(frame_files)-1, num_samples, dtype=int)

    plt.figure(figsize=(15, 3*num_samples))

    for i, idx in enumerate(sample_indices):
        frame_path = os.path.join(frames_dir, frame_files[idx])
        pred_score = predictions[idx]

        # Load and display the frame
        img = cv2.imread(frame_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

        plt.subplot(num_samples, 1, i+1)
        plt.imshow(img)

        # Use 0.3 as the threshold for categorization
        if pred_score > 0.3:
            color = 'red'  # Fake
            status = "FAKE"
        else:
            color = 'green'  # Real
            status = "REAL"

        plt.title(f"Frame {idx} - Score: {pred_score:.4f} ({status})", color=color, fontweight='bold')
        plt.axis('off')

    plt.tight_layout()

    save_path = '/content/sample_frames_analysis.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/sample_frames_analysis.png'

    plt.savefig(save_path)
    plt.show()

    print("📊 Sample frames analysis saved")

# Main function to run the entire process
def detect_deepfake():
    """
    Complete workflow for deepfake detection:
    1. Load model
    2. Upload video
    3. Analyze video
    4. Display results
    """
    print("\n🔍 DEEPFAKE DETECTION TOOL")
    print("=" * 50)

    # Step 1: Load the model
    model = load_deepfake_model()
    if model is None:
        print("❌ Failed to load model. Exiting.")
        return

    # Step 2: Get video
    print("\n📹 Upload a video to analyze")
    uploaded = files.upload()

    if not uploaded:
        print("❌ No video was uploaded. Exiting.")
        return

    video_filename = list(uploaded.keys())[0]
    video_path = os.path.join(os.getcwd(), video_filename)

    # Step 3: Create temp directory for frames
    temp_frames_dir = "/content/temp_frames"
    if os.path.exists(temp_frames_dir):
        shutil.rmtree(temp_frames_dir)
    os.makedirs(temp_frames_dir)

    # Step 4: Analyze the video
    try:
        print(f"\n🔍 Analyzing video: {video_filename}")
        print("This may take a few minutes depending on the video length...")

        start_time = time.time()
        results = analyze_video(model, video_path, temp_frames_dir)
        elapsed_time = time.time() - start_time

        # Step 5: Display results
        print_results(results)
        plot_sample_frames(temp_frames_dir, results['frame_predictions'])

        print(f"\n⏱️ Analysis completed in {elapsed_time:.2f} seconds")

        # Step 6: Save detailed report
        if os.path.exists('/content/drive'):
            save_detailed_report(video_filename, results, elapsed_time)

    except Exception as e:
        print(f"❌ Error analyzing video: {e}")

    finally:
        # Clean up temp files
        if os.path.exists(temp_frames_dir):
            shutil.rmtree(temp_frames_dir)
            print("🧹 Temporary files cleaned up")

# Function to save a detailed report
def save_detailed_report(video_name, results, elapsed_time):
    """
    Save a detailed analysis report to Google Drive

    Args:
        video_name: Name of the analyzed video
        results: Dictionary containing analysis results
        elapsed_time: Time taken for analysis (in seconds)
    """
    report_path = "/content/drive/MyDrive/deepfake_analysis_reports"
    os.makedirs(report_path, exist_ok=True)

    # Create timestamp
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    report_file = os.path.join(report_path, f"analysis_{timestamp}_{video_name}.txt")

    with open(report_file, 'w') as f:
        f.write("=" * 60 + "\n")
        f.write(f"DEEPFAKE DETECTION REPORT - {timestamp}\n")
        f.write("=" * 60 + "\n\n")

        f.write(f"Video analyzed: {video_name}\n")
        f.write(f"Analysis date: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Analysis duration: {elapsed_time:.2f} seconds\n\n")

        f.write("RESULTS SUMMARY:\n")
        f.write("-" * 30 + "\n")
        f.write(f"Total frames analyzed: {results['total_frames']}\n")
        f.write(f"Frames detected as fake: {results['fake_count']}\n")
        f.write(f"Fake percentage: {results['fake_percentage']:.2f}%\n")
        f.write(f"Average prediction score: {results['avg_score']:.4f}\n")
        f.write(f"Confidence level: {results['confidence']:.2f}\n\n")

        if results['is_fake']:
            f.write("VERDICT: FAKE\n")
        else:
            f.write("VERDICT: REAL\n")

        if results['confidence'] < 0.3:
            f.write("Note: Low confidence detection\n")
        elif results['confidence'] > 0.7:
            f.write("High confidence detection\n")

        f.write("\nDETAILED FRAME ANALYSIS:\n")
        f.write("-" * 30 + "\n")

        # Update frame-level statistics using 0.3 threshold
        frames_fake = sum(1 for p in results['frame_predictions'] if p > 0.3)
        frames_real = sum(1 for p in results['frame_predictions'] if p <= 0.3)

        f.write(f"FAKE frames (score > 0.3): {frames_fake} frames\n")
        f.write(f"REAL frames (score <= 0.3): {frames_real} frames\n\n")

        f.write("First 20 frame predictions:\n")
        for i, pred in enumerate(results['frame_predictions'][:20]):
            status = "FAKE" if pred > 0.3 else "REAL"
            f.write(f"Frame {i}: {pred:.4f} - {status}\n")

        if len(results['frame_predictions']) > 20:
            f.write("...\n")

    print(f"✅ Detailed analysis report saved to: {report_file}")

# Just run this function to use your deepfake detection model
if __name__ == "__main__":
    detect_deepfake()
```

```python
# Enhanced deepfake detection with background analysis
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
from google.colab import drive, files
from tqdm import tqdm
import shutil
import time

# Mount Google Drive where your model is stored
drive.mount('/content/drive')

def load_deepfake_model(model_path=None):
    """
    Load the saved deepfake detection model

    Args:
        model_path: Path to the saved model (.h5 file)
                   If None, will prompt for file upload

    Returns:
        Loaded Keras model
    """
    if model_path is None or not os.path.exists(model_path):
        # Ask if model is in Google Drive or needs to be uploaded
        print("Select how to load your model:")
        print("1. Load from Google Drive")
        print("2. Upload model file")

        choice = input("Enter your choice (1 or 2): ")

        if choice == "1":
            # Search for model files in Google Drive
            print("\nSearching for model files in Google Drive...")
            drive_path = "/content/drive/MyDrive"
            model_files = [f for f in os.listdir(drive_path) if f.endswith('.h5')]

            if not model_files:
                print("❌ No .h5 model files found in Google Drive root.")
                print("📂 Please specify the exact path or upload a model file.")
                return None

            print("\nFound these model files in Google Drive:")
            for i, f in enumerate(model_files):
                print(f"{i+1}. {f}")

            idx = int(input("\nEnter the number of the model to load: ")) - 1
            if 0 <= idx < len(model_files):
                model_path = os.path.join(drive_path, model_files[idx])
            else:
                print("❌ Invalid selection.")
                return None
        else:
            # Upload model file
            print("\n📤 Please upload your .h5 model file...")
            uploaded = files.upload()

            if not uploaded:
                print("❌ No file was uploaded.")
                return None

            model_path = list(uploaded.keys())[0]

    # Load the model
    print(f"🔄 Loading model from {model_path}...")
    try:
        model = load_model(model_path)
        print("✅ Model loaded successfully!")
        return model
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return None

# Function to extract frames from video
def extract_frames(video_path, output_folder, frames_per_second=1):
    """
    Extract frames from a video file.

    Args:
        video_path: Path to the video file
        output_folder: Folder to save the frames
        frames_per_second: Number of frames to extract per second

    Returns:
        List of paths to extracted frames
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidcap = cv2.VideoCapture(video_path)
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Extract frame every nth frame
    frame_interval = int(fps / frames_per_second)

    frame_count = 0
    extracted_count = 0
    frame_paths = []

    print(f"Extracting frames from video (total frames: {total_frames})")

    while True:
        success, frame = vidcap.read()
        if not success:
            break

        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{extracted_count:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_paths.append(frame_path)
            extracted_count += 1

        frame_count += 1

        # Show progress
        if frame_count % 100 == 0:
            print(f"Processed {frame_count}/{total_frames} frames")

    vidcap.release()
    print(f"Extracted {extracted_count} frames to {output_folder}")
    return frame_paths

# NEW FUNCTION: Detection of faces in images
def detect_faces(image):
    """
    Detects faces in an image using OpenCV's pre-trained face detector

    Args:
        image: Input image (numpy array)

    Returns:
        List of face bounding boxes (x, y, w, h)
    """
    # Load OpenCV's pre-trained face detector
    face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
    face_cascade = cv2.CascadeClassifier(face_cascade_path)

    # Convert image to grayscale for face detection
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    return faces

# NEW FUNCTION: ELA (Error Level Analysis) - detects compression artifacts
def error_level_analysis(image_path, quality=90):
    """
    Performs Error Level Analysis on an image to detect manipulation

    Args:
        image_path: Path to the image
        quality: JPEG compression quality

    Returns:
        ELA image showing potential manipulation areas
    """
    # Read the image
    original = cv2.imread(image_path)

    # Save with specific quality to temporary file
    temp_path = image_path + '.temp.jpg'
    cv2.imwrite(temp_path, original, [cv2.IMWRITE_JPEG_QUALITY, quality])

    # Read the temp file
    compressed = cv2.imread(temp_path)

    # Calculate the difference
    diff = cv2.absdiff(original, compressed)

    # Enhance the difference for visualization
    diff_enhanced = cv2.convertScaleAbs(diff, alpha=10, beta=0)

    # Remove temporary file
    os.remove(temp_path)

    return diff_enhanced

# NEW FUNCTION: Analyze background noise consistency
def analyze_background_noise(image):
    """
    Analyzes background noise patterns of an image for consistency

    Args:
        image: Input image

    Returns:
        Background noise score (higher = more suspicious)
    """
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce details
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Laplacian filter to detect edges and noise patterns
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

    # Calculate standard deviation of Laplacian as a measure of noise consistency
    noise_score = np.std(laplacian)

    # Normalize to 0-1 range (based on empirical observation)
    normalized_score = min(1.0, noise_score / 20.0)

    return normalized_score

# NEW FUNCTION: Check for temporal consistency between frames
def check_temporal_consistency(frame1, frame2):
    """
    Checks for temporal consistency between consecutive frames

    Args:
        frame1: First frame
        frame2: Second frame

    Returns:
        Consistency score (lower = more consistent)
    """
    # Convert to grayscale
    gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Calculate optical flow
    flow = cv2.calcOpticalFlowFarneback(
        gray1, gray2, None, 0.5, 3, 15, 3, 5, 1.2, 0
    )

    # Calculate magnitude and angle of flow vectors
    mag, _ = cv2.cartToPolar(flow[..., 0], flow[..., 1])

    # Calculate statistics of magnitude
    mean_mag = np.mean(mag)
    std_mag = np.std(mag)

    # Higher variance in flow might indicate inconsistency
    # Normalize to 0-1 range
    consistency_score = min(1.0, std_mag / (mean_mag + 1e-5) / 5.0)

    return consistency_score

# NEW FUNCTION: Background segmentation
def segment_background(image, face_regions):
    """
    Segments the background from the image by masking out face regions

    Args:
        image: Input image
        face_regions: List of face bounding boxes (x, y, w, h)

    Returns:
        Background mask, background image
    """
    # Create a mask with the same size as the image
    mask = np.ones(image.shape[:2], dtype=np.uint8) * 255

    # Mask out face regions
    for (x, y, w, h) in face_regions:
        # Add some margin to the face region
        margin = int(min(w, h) * 0.2)
        x_start = max(0, x - margin)
        y_start = max(0, y - margin)
        x_end = min(image.shape[1], x + w + margin)
        y_end = min(image.shape[0], y + h + margin)

        # Set face region to 0 in the mask
        mask[y_start:y_end, x_start:x_end] = 0

    # Extract background using the mask
    background = cv2.bitwise_and(image, image, mask=mask)

    return mask, background

# Enhanced function to analyze frames in batches with background analysis
def batch_analyze_frames(model, frame_paths, batch_size=16):
    """
    Process frames in batches to avoid memory issues with large videos
    Includes additional analysis for backgrounds and temporal consistency

    Args:
        model: Trained deepfake detection model
        frame_paths: List of paths to video frames
        batch_size: Number of frames to process at once

    Returns:
        Dictionary with prediction scores and additional analysis results
    """
    from tensorflow.keras.preprocessing import image

    all_predictions = []
    background_scores = []
    temporal_scores = []
    ela_scores = []
    previous_frame = None

    # Process frames in batches
    for i in tqdm(range(0, len(frame_paths), batch_size)):
        batch_paths = frame_paths[i:i+batch_size]
        batch_images = []
        batch_original_images = []

        # Prepare batch
        for frame_path in batch_paths:
            # Load image for model prediction
            img = image.load_img(frame_path, target_size=(224, 224))
            img_array = image.img_to_array(img) / 255.0
            batch_images.append(img_array)

            # Load original image for additional analysis
            original_img = cv2.imread(frame_path)
            batch_original_images.append(original_img)

            # Perform ELA analysis
            ela_img = error_level_analysis(frame_path)
            ela_score = np.mean(ela_img) / 255.0
            ela_scores.append(ela_score)

        # Convert to numpy array for model prediction
        batch_data = np.array(batch_images)

        # Get model predictions
        batch_preds = model.predict(batch_data, verbose=0)

        # Process each original image for additional analysis
        for j, original_img in enumerate(batch_original_images):
            # Detect faces
            faces = detect_faces(original_img)

            # Analyze background
            if len(faces) > 0:
                # If faces detected, segment and analyze the background
                _, background = segment_background(original_img, faces)
                bg_score = analyze_background_noise(background)
            else:
                # If no faces detected, analyze the whole image
                bg_score = analyze_background_noise(original_img)

            background_scores.append(bg_score)

            # Check temporal consistency if we have a previous frame
            if previous_frame is not None:
                temp_score = check_temporal_consistency(previous_frame, original_img)
                temporal_scores.append(temp_score)
            else:
                # For the first frame, assign a neutral score
                if j > 0 or i > 0:  # Not the very first frame in the video
                    temp_score = check_temporal_consistency(batch_original_images[j-1], original_img)
                    temporal_scores.append(temp_score)
                else:
                    temporal_scores.append(0.5)  # Neutral score for first frame

            # Update previous frame
            previous_frame = original_img.copy()

        # Store model predictions
        all_predictions.extend([float(p[0]) for p in batch_preds])

    # Calculate combined scores
    combined_scores = []
    for i in range(len(all_predictions)):
        # Ensure we have scores for all frames
        bg_score = background_scores[i] if i < len(background_scores) else 0.5
        temp_score = temporal_scores[i] if i < len(temporal_scores) else 0.5
        ela_score = ela_scores[i] if i < len(ela_scores) else 0.5

        # Weighted combination (can be adjusted)
        model_weight = 0.6  # Main model has higher weight
        bg_weight = 0.15    # Background analysis weight
        temp_weight = 0.15  # Temporal consistency weight
        ela_weight = 0.1    # ELA analysis weight

        combined_score = (all_predictions[i] * model_weight +
                         bg_score * bg_weight +
                         temp_score * temp_weight +
                         ela_score * ela_weight)

        # Ensure score is between 0 and 1
        combined_score = max(0, min(1, combined_score))
        combined_scores.append(combined_score)

    return {
        'model_predictions': all_predictions,
        'background_scores': background_scores,
        'temporal_scores': temporal_scores,
        'ela_scores': ela_scores,
        'combined_scores': combined_scores
    }

# Enhanced function to analyze a video
def analyze_video(model, video_path, temp_folder="/content/temp_frames", batch_size=16):
    """
    Analyze a video for deepfake detection with enhanced background analysis.

    Args:
        model: Trained model
        video_path: Path to the video
        temp_folder: Folder to save temporary frames
        batch_size: Batch size for processing frames

    Returns:
        Dictionary with analysis results
    """
    # Extract frames
    frame_paths = extract_frames(video_path, temp_folder)

    if not frame_paths:
        raise ValueError("No frames could be extracted from the video")

    # Analyze frames in batches with enhanced analysis
    print("\nAnalyzing frames with enhanced background and temporal analysis...")
    analysis_results = batch_analyze_frames(model, frame_paths, batch_size)

    # Use the combined scores for final determination
    frame_predictions = analysis_results['combined_scores']

    # Calculate statistics
    avg_score = np.mean(frame_predictions)

    # Use 0.3 as the threshold, as requested
    threshold = 0.3

    binary_preds = [1 if p > threshold else 0 for p in frame_predictions]
    fake_count = sum(binary_preds)
    total = len(binary_preds)
    fake_percentage = (fake_count / total) * 100

    # Get confidence level
    confidence = abs(avg_score - 0.5) * 2  # Scale to 0-1

    # Plot distribution of different scores
    plt.figure(figsize=(15, 10))

    # Plot model predictions
    plt.subplot(2, 2, 1)
    plt.hist(analysis_results['model_predictions'], bins=10, alpha=0.7, color='blue')
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Model Prediction Scores')
    plt.xlabel('Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()

    # Plot background scores
    plt.subplot(2, 2, 2)
    plt.hist(analysis_results['background_scores'], bins=10, alpha=0.7, color='green')
    plt.title('Background Analysis Scores')
    plt.xlabel('Score (Higher = More Suspicious)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)

    # Plot temporal consistency scores
    plt.subplot(2, 2, 3)
    plt.hist(analysis_results['temporal_scores'], bins=10, alpha=0.7, color='orange')
    plt.title('Temporal Consistency Scores')
    plt.xlabel('Score (Higher = Less Consistent)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)

    # Plot combined scores
    plt.subplot(2, 2, 4)
    plt.hist(frame_predictions, bins=10, alpha=0.7, color='purple')
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Combined Scores (Final Decision)')
    plt.xlabel('Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.tight_layout()

    save_path = '/content/score_distributions.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/score_distributions.png'

    plt.savefig(save_path)
    plt.show()

    # Create second plot showing frame-by-frame scores
    plt.figure(figsize=(15, 8))
    frame_indices = range(len(frame_predictions))

    plt.plot(frame_indices, analysis_results['model_predictions'], 'b-', alpha=0.6, label='Model Score')
    plt.plot(frame_indices, analysis_results['background_scores'], 'g-', alpha=0.6, label='Background Score')
    plt.plot(frame_indices, analysis_results['temporal_scores'], 'y-', alpha=0.6, label='Temporal Score')
    plt.plot(frame_indices, frame_predictions, 'r-', linewidth=2, label='Combined Score')

    plt.axhline(y=threshold, color='purple', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Frame-by-Frame Analysis Scores')
    plt.xlabel('Frame Number')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_path2 = '/content/frame_analysis.png'
    if os.path.exists('/content/drive'):
        save_path2 = '/content/drive/MyDrive/frame_analysis.png'

    plt.savefig(save_path2)
    plt.show()

    return {
        'model_predictions': analysis_results['model_predictions'],
        'background_scores': analysis_results['background_scores'],
        'temporal_scores': analysis_results['temporal_scores'],
        'ela_scores': analysis_results['ela_scores'],
        'frame_predictions': frame_predictions,  # These are the combined scores
        'fake_count': fake_count,
        'total_frames': total,
        'fake_percentage': fake_percentage,
        'avg_score': avg_score,
        'confidence': confidence,
        'is_fake': fake_percentage > 50
    }

# Enhanced results printing function
def print_results(results):
    """Print the enhanced results of video analysis"""
    print("\n" + "="*60)
    print("🔍 ENHANCED DEEPFAKE DETECTION RESULTS 🔍")
    print("="*60)

    print(f"🧠 Detected {results['fake_count']}/{results['total_frames']} frames as FAKE")
    print(f"📊 Fake percentage: {results['fake_percentage']:.2f}%")
    print(f"⚖️ Average prediction score: {results['avg_score']:.4f}")
    print(f"🎯 Confidence level: {results['confidence']:.2f}")

    # Calculate average scores for each component
    avg_model = np.mean(results['model_predictions'])
    avg_bg = np.mean(results['background_scores'])
    avg_temp = np.mean(results['temporal_scores'])
    avg_ela = np.mean(results['ela_scores'])

    print("\n📈 DETECTION COMPONENT SCORES:")
    print(f"   🔹 Base model score: {avg_model:.4f}")
    print(f"   🔹 Background analysis: {avg_bg:.4f}")
    print(f"   🔹 Temporal consistency: {avg_temp:.4f}")
    print(f"   🔹 Error Level Analysis: {avg_ela:.4f}")

    # Identify which components contributed most to the detection
    scores = [
        ("Base model", avg_model),
        ("Background analysis", avg_bg),
        ("Temporal consistency", avg_temp),
        ("Error level analysis", avg_ela)
    ]

    # Sort by highest score (most suspicious)
    scores.sort(key=lambda x: x[1], reverse=True)

    print("\n🔎 TOP DETECTION SIGNALS:")
    for i, (name, score) in enumerate(scores[:2]):
        if score > 0.4:  # Only show significant signals
            print(f"   {i+1}. {name}: {score:.4f}")

    if results['is_fake']:
        print("\n🔴 VERDICT: This video is classified as FAKE")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")
    else:
        print("\n🟢 VERDICT: This video is classified as REAL")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")

    print("="*60)
    print("📈 Check the saved distribution charts for more details")
    print("="*60)

# Enhanced function to visualize sample frames with their predictions
def plot_sample_frames(frames_dir, results, num_samples=5):
    """
    Plot sample frames from the video with their prediction scores and analysis

    Args:
        frames_dir: Directory containing video frames
        results: Dictionary with analysis results
        num_samples: Number of sample frames to show
    """
    import cv2
    import numpy as np

    # Get all frame files
    frame_files = sorted([f for f in os.listdir(frames_dir) if f.endswith('.jpg')])

    if len(frame_files) == 0:
        print("No frames found to display")
        return

    # Select sample frames (evenly distributed)
    if len(frame_files) <= num_samples:
        sample_indices = list(range(len(frame_files)))
    else:
        sample_indices = np.linspace(0, len(frame_files)-1, num_samples, dtype=int)

    plt.figure(figsize=(18, 4*num_samples))

    for i, idx in enumerate(sample_indices):
        frame_path = os.path.join(frames_dir, frame_files[idx])

        # Get all scores for this frame
        combined_score = results['frame_predictions'][idx]
        model_score = results['model_predictions'][idx]
        bg_score = results['background_scores'][idx]
        temp_score = results['temporal_scores'][idx] if idx < len(results['temporal_scores']) else 0
        ela_score = results['ela_scores'][idx]

        # Load and display the frame
        img = cv2.imread(frame_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

        # Perform ELA on the frame for visualization
        ela_img = error_level_analysis(frame_path)
        ela_img = cv2.cvtColor(ela_img, cv2.COLOR_BGR2RGB)

        # Detect faces to visualize segmentation
        faces = detect_faces(img)
        img_with_faces = img.copy()

        # Draw rectangles around faces
        for (x, y, w, h) in faces:
            cv2.rectangle(img_with_faces, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Get background mask
        bg_mask, background = segment_background(img, faces)

        # Create a subplot for this frame with 4 images
        plt.subplot(num_samples, 4, i*4+1)
        plt.imshow(img_with_faces)
        plt.title(f"Frame {idx}\nScore: {combined_score:.3f}",
                 color='red' if combined_score > 0.3 else 'green',
                 fontweight='bold')
        plt.axis('off')

        plt.subplot(num_samples, 4, i*4+2)
        plt.imshow(background)
        plt.title(f"Background\nScore: {bg_score:.3f}",
                 color='red' if bg_score > 0.5 else 'green')
        plt.axis('off')

        plt.subplot(num_samples, 4, i*4+3)
        plt.imshow(ela_img)
        plt.title(f"Error Level Analysis\nScore: {ela_score:.3f}",
                 color='red' if ela_score > 0.5 else 'green')
        plt.axis('off')

        # Plot a mini-dashboard of scores
        plt.subplot(num_samples, 4, i*4+4)
        score_labels = ['Combined', 'Model', 'Background', 'Temporal', 'ELA']
        score_values = [combined_score, model_score, bg_score, temp_score, ela_score]
        colors = ['red' if s > 0.3 else 'green' for s in score_values]

        y_pos = np.arange(len(score_labels))
        plt.barh(y_pos, score_values, color=colors)
        plt.yticks(y_pos, score_labels)
        plt.xlim(0, 1)
        plt.title("Component Scores")
        plt.grid(axis='x', alpha=0.3)

    plt.tight_layout()

    save_path = '/content/enhanced_sample_frames_analysis.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/enhanced_sample_frames_analysis.png'

    plt.savefig(save_path)
    plt.show()

    print("📊 Enhanced sample frames analysis saved")

# Enhanced function to save a detailed report
def save_detailed_report(video_name, results, elapsed_time):
    """
    Save a detailed analysis report to Google Drive

    Args:
        video_name: Name of the analyzed video
        results: Dictionary containing analysis results
        elapsed_time: Time taken for analysis (in seconds)
    """
    report_path = "/content/drive/MyDrive/deepfake_analysis_reports"
    os.makedirs(report_path, exist_ok=True)

    # Create timestamp
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    report_file = os.path.join(report_path, f"enhanced_analysis_{timestamp}_{video_name}.txt")

    with open(report_file, 'w') as f:
        f.write("=" * 70 + "\n")
        f.write(f"ENHANCED DEEPFAKE DETECTION REPORT - {timestamp}\n")
        f.write("=" * 70 + "\n\n")

        f.write(f"Video analyzed: {video_name}\n")
        f.write(f"Analysis date: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Analysis duration: {elapsed_time:.2f} seconds\n\n")

        f.write("RESULTS SUMMARY:\n")
        f.write("-" * 30 + "\n")
        f.write(f"Total frames analyzed: {results['total_frames']}\n")
        f.write(f"Frames detected as fake: {results['fake_count']}\n")
        f.write(f"Fake percentage: {results['fake_percentage']:.2f}%\n")
        f.write(f"Average prediction score: {results['avg_score']:.4f}\n")
        f.write(f"Confidence level: {results['confidence']:.2f}\n\n")

        # Add component analysis
        f.write("COMPONENT ANALYSIS:\n")
        f.write("-" * 30 + "\n")

        avg_model = np.mean(results['model_predictions'])
        avg_bg = np.mean(results['background_scores'])
        avg_temp = np.mean(results['temporal_scores'])
        avg_ela = np.mean(results['ela_scores'])

        f.write(f"Base model average score: {avg_model:.4f}\n")
        f.write(f"Background analysis average score: {avg_bg:.4f}\n")
        f.write(f"Temporal consistency average score: {avg_temp:.4f}\n")
        f.write(f"Error Level Analysis average score: {avg_ela:.4f}\n\n")

        if results['is_fake']:
            f.write("VERDICT: FAKE\n")
        else:
            f.write("VERDICT: REAL\n")

        if results['confidence'] < 0.3:
            f.write("Note: Low confidence detection\n")
        elif results['confidence'] > 0.7:
            f.write("High confidence detection\n")

        f.write("\nDETAILED FRAME ANALYSIS:\n")
        f.write("-" * 30 + "\n")

        # Update frame-level statistics using 0.3 threshold
        frames_fake = sum(1 for p in results['frame_predictions'] if p > 0.3)
        frames_real = sum(1 for p in results['frame_predictions'] if p <= 0.3)

        f.write(f"FAKE frames (score > 0.3): {frames_fake} frames\n")
        f.write(f"REAL frames (score <= 0.3): {frames_real} frames\n\n")

        f.write("First 20 frame predictions with component breakdown:\n")
        f.write("Frame\tCombined\tModel\tBackground\tTemporal\tELA\tVerdict\n")

        # Write the first 20 frames details
        num_frames_to_show = min(20, len(results['frame_predictions']))
        for i in range(num_frames_to_show):
            combined = results['frame_predictions'][i]
            model = results['model_predictions'][i]
            bg = results['background_scores'][i] if i < len(results['background_scores']) else 0
            temp = results['temporal_scores'][i] if i < len(results['temporal_scores']) else 0
            ela = results['ela_scores'][i] if i < len(results['ela_scores']) else 0

            verdict = "FAKE" if combined > 0.3 else "REAL"

            f.write(f"{i}\t{combined:.3f}\t{model:.3f}\t{bg:.3f}\t{temp:.3f}\t{ela:.3f}\t{verdict}\n")

        f.write("\nANALYSIS INTERPRETATION:\n")
        f.write("-" * 30 + "\n")

        # Interpretation based on component scores
        if avg_model > 0.5:
            f.write("- The base deepfake detection model strongly indicates manipulation\n")

        if avg_bg > 0.5:
            f.write("- Background inconsistencies detected, suggesting possible compositing\n")

        if avg_temp > 0.5:
            f.write("- Temporal inconsistencies detected between frames\n")

        if avg_ela > 0.5:
            f.write("- Error Level Analysis shows compression artifacts consistent with manipulation\n")

        # Overall assessment
        f.write("\nOVERALL ASSESSMENT:\n")
        f.write("-" * 30 + "\n")

        if results['is_fake']:
            if results['confidence'] > 0.7:
                f.write("This video shows strong evidence of being manipulated or generated.\n")
            elif results['confidence'] > 0.3:
                f.write("This video shows moderate evidence of manipulation, but further analysis may be needed.\n")
            else:
                f.write("This video shows weak indicators of manipulation, but results are inconclusive.\n")
        else:
            if results['confidence'] > 0.7:
                f.write("This video shows strong indicators of being authentic.\n")
            elif results['confidence'] > 0.3:
                f.write("This video appears authentic, but with moderate confidence.\n")
            else:
                f.write("This video lacks strong manipulation indicators, but results are inconclusive.\n")

        f.write("\n" + "=" * 70 + "\n")
        f.write("End of report\n")
        f.write("=" * 70)

    print(f"📝 Detailed report saved to {report_file}")
if __name__ == "__main__":
   detect_deepfake()
```

```python
# Enhanced deepfake detection with background analysis
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
from google.colab import drive, files
from tqdm import tqdm
import shutil
import time

# Mount Google Drive where your model is stored
drive.mount('/content/drive')

def load_deepfake_model(model_path=None):
    """
    Load the saved deepfake detection model

    Args:
        model_path: Path to the saved model (.h5 file)
                   If None, will prompt for file upload

    Returns:
        Loaded Keras model
    """
    if model_path is None or not os.path.exists(model_path):
        # Ask if model is in Google Drive or needs to be uploaded
        print("Select how to load your model:")
        print("1. Load from Google Drive")
        print("2. Upload model file")

        choice = input("Enter your choice (1 or 2): ")

        if choice == "1":
            # Search for model files in Google Drive
            print("\nSearching for model files in Google Drive...")
            drive_path = "/content/drive/MyDrive"
            model_files = [f for f in os.listdir(drive_path) if f.endswith('.h5')]

            if not model_files:
                print("❌ No .h5 model files found in Google Drive root.")
                print("📂 Please specify the exact path or upload a model file.")
                return None

            print("\nFound these model files in Google Drive:")
            for i, f in enumerate(model_files):
                print(f"{i+1}. {f}")

            idx = int(input("\nEnter the number of the model to load: ")) - 1
            if 0 <= idx < len(model_files):
                model_path = os.path.join(drive_path, model_files[idx])
            else:
                print("❌ Invalid selection.")
                return None
        else:
            # Upload model file
            print("\n📤 Please upload your .h5 model file...")
            uploaded = files.upload()

            if not uploaded:
                print("❌ No file was uploaded.")
                return None

            model_path = list(uploaded.keys())[0]

    # Load the model
    print(f"🔄 Loading model from {model_path}...")
    try:
        model = load_model(model_path)
        print("✅ Model loaded successfully!")
        return model
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return None

# Function to extract frames from video
def extract_frames(video_path, output_folder, frames_per_second=1):
    """
    Extract frames from a video file.

    Args:
        video_path: Path to the video file
        output_folder: Folder to save the frames
        frames_per_second: Number of frames to extract per second

    Returns:
        List of paths to extracted frames
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidcap = cv2.VideoCapture(video_path)
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Extract frame every nth frame
    frame_interval = int(fps / frames_per_second)

    frame_count = 0
    extracted_count = 0
    frame_paths = []

    print(f"Extracting frames from video (total frames: {total_frames})")

    while True:
        success, frame = vidcap.read()
        if not success:
            break

        if frame_count % frame_interval == 0:
            frame_path = os.path.join(output_folder, f"frame_{extracted_count:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_paths.append(frame_path)
            extracted_count += 1

        frame_count += 1

        # Show progress
        if frame_count % 100 == 0:
            print(f"Processed {frame_count}/{total_frames} frames")

    vidcap.release()
    print(f"Extracted {extracted_count} frames to {output_folder}")
    return frame_paths

# NEW FUNCTION: Detection of faces in images
def detect_faces(image):
    """
    Detects faces in an image using OpenCV's pre-trained face detector

    Args:
        image: Input image (numpy array)

    Returns:
        List of face bounding boxes (x, y, w, h)
    """
    # Load OpenCV's pre-trained face detector
    face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
    face_cascade = cv2.CascadeClassifier(face_cascade_path)

    # Convert image to grayscale for face detection
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    return faces

# NEW FUNCTION: ELA (Error Level Analysis) - detects compression artifacts
def error_level_analysis(image_path, quality=90):
    """
    Performs Error Level Analysis on an image to detect manipulation

    Args:
        image_path: Path to the image
        quality: JPEG compression quality

    Returns:
        ELA image showing potential manipulation areas
    """
    # Read the image
    original = cv2.imread(image_path)

    # Save with specific quality to temporary file
    temp_path = image_path + '.temp.jpg'
    cv2.imwrite(temp_path, original, [cv2.IMWRITE_JPEG_QUALITY, quality])

    # Read the temp file
    compressed = cv2.imread(temp_path)

    # Calculate the difference
    diff = cv2.absdiff(original, compressed)

    # Enhance the difference for visualization
    diff_enhanced = cv2.convertScaleAbs(diff, alpha=10, beta=0)

    # Remove temporary file
    os.remove(temp_path)

    return diff_enhanced

# NEW FUNCTION: Analyze background noise consistency
def analyze_background_noise(image):
    """
    Analyzes background noise patterns of an image for consistency

    Args:
        image: Input image

    Returns:
        Background noise score (higher = more suspicious)
    """
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce details
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Laplacian filter to detect edges and noise patterns
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

    # Calculate standard deviation of Laplacian as a measure of noise consistency
    noise_score = np.std(laplacian)

    # Normalize to 0-1 range (based on empirical observation)
    normalized_score = min(1.0, noise_score / 20.0)

    return normalized_score

# NEW FUNCTION: Check for temporal consistency between frames
def check_temporal_consistency(frame1, frame2):
    """
    Checks for temporal consistency between consecutive frames

    Args:
        frame1: First frame
        frame2: Second frame

    Returns:
        Consistency score (lower = more consistent)
    """
    # Convert to grayscale
    gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Calculate optical flow
    flow = cv2.calcOpticalFlowFarneback(
        gray1, gray2, None, 0.5, 3, 15, 3, 5, 1.2, 0
    )

    # Calculate magnitude and angle of flow vectors
    mag, _ = cv2.cartToPolar(flow[..., 0], flow[..., 1])

    # Calculate statistics of magnitude
    mean_mag = np.mean(mag)
    std_mag = np.std(mag)

    # Higher variance in flow might indicate inconsistency
    # Normalize to 0-1 range
    consistency_score = min(1.0, std_mag / (mean_mag + 1e-5) / 5.0)

    return consistency_score

# NEW FUNCTION: Background segmentation
def segment_background(image, face_regions):
    """
    Segments the background from the image by masking out face regions

    Args:
        image: Input image
        face_regions: List of face bounding boxes (x, y, w, h)

    Returns:
        Background mask, background image
    """
    # Create a mask with the same size as the image
    mask = np.ones(image.shape[:2], dtype=np.uint8) * 255

    # Mask out face regions
    for (x, y, w, h) in face_regions:
        # Add some margin to the face region
        margin = int(min(w, h) * 0.2)
        x_start = max(0, x - margin)
        y_start = max(0, y - margin)
        x_end = min(image.shape[1], x + w + margin)
        y_end = min(image.shape[0], y + h + margin)

        # Set face region to 0 in the mask
        mask[y_start:y_end, x_start:x_end] = 0

    # Extract background using the mask
    background = cv2.bitwise_and(image, image, mask=mask)

    return mask, background

# Enhanced function to analyze frames in batches with background analysis
def batch_analyze_frames(model, frame_paths, batch_size=16):
    """
    Process frames in batches to avoid memory issues with large videos
    Includes additional analysis for backgrounds and temporal consistency

    Args:
        model: Trained deepfake detection model
        frame_paths: List of paths to video frames
        batch_size: Number of frames to process at once

    Returns:
        Dictionary with prediction scores and additional analysis results
    """
    from tensorflow.keras.preprocessing import image

    all_predictions = []
    background_scores = []
    temporal_scores = []
    ela_scores = []
    previous_frame = None

    # Process frames in batches
    for i in tqdm(range(0, len(frame_paths), batch_size)):
        batch_paths = frame_paths[i:i+batch_size]
        batch_images = []
        batch_original_images = []

        # Prepare batch
        for frame_path in batch_paths:
            # Load image for model prediction
            img = image.load_img(frame_path, target_size=(224, 224))
            img_array = image.img_to_array(img) / 255.0
            batch_images.append(img_array)

            # Load original image for additional analysis
            original_img = cv2.imread(frame_path)
            batch_original_images.append(original_img)

            # Perform ELA analysis
            ela_img = error_level_analysis(frame_path)
            ela_score = np.mean(ela_img) / 255.0
            ela_scores.append(ela_score)

        # Convert to numpy array for model prediction
        batch_data = np.array(batch_images)

        # Get model predictions
        batch_preds = model.predict(batch_data, verbose=0)

        # Process each original image for additional analysis
        for j, original_img in enumerate(batch_original_images):
            # Detect faces
            faces = detect_faces(original_img)

            # Analyze background
            if len(faces) > 0:
                # If faces detected, segment and analyze the background
                _, background = segment_background(original_img, faces)
                bg_score = analyze_background_noise(background)
            else:
                # If no faces detected, analyze the whole image
                bg_score = analyze_background_noise(original_img)

            background_scores.append(bg_score)

            # Check temporal consistency if we have a previous frame
            if previous_frame is not None:
                temp_score = check_temporal_consistency(previous_frame, original_img)
                temporal_scores.append(temp_score)
            else:
                # For the first frame, assign a neutral score
                if j > 0 or i > 0:  # Not the very first frame in the video
                    temp_score = check_temporal_consistency(batch_original_images[j-1], original_img)
                    temporal_scores.append(temp_score)
                else:
                    temporal_scores.append(0.5)  # Neutral score for first frame

            # Update previous frame
            previous_frame = original_img.copy()

        # Store model predictions
        all_predictions.extend([float(p[0]) for p in batch_preds])

    # Calculate combined scores
    combined_scores = []
    for i in range(len(all_predictions)):
        # Ensure we have scores for all frames
        bg_score = background_scores[i] if i < len(background_scores) else 0.5
        temp_score = temporal_scores[i] if i < len(temporal_scores) else 0.5
        ela_score = ela_scores[i] if i < len(ela_scores) else 0.5

        # Weighted combination (can be adjusted)
        model_weight = 0.6  # Main model has higher weight
        bg_weight = 0.15    # Background analysis weight
        temp_weight = 0.15  # Temporal consistency weight
        ela_weight = 0.1    # ELA analysis weight

        combined_score = (all_predictions[i] * model_weight +
                         bg_score * bg_weight +
                         temp_score * temp_weight +
                         ela_score * ela_weight)

        # Ensure score is between 0 and 1
        combined_score = max(0, min(1, combined_score))
        combined_scores.append(combined_score)

    return {
        'model_predictions': all_predictions,
        'background_scores': background_scores,
        'temporal_scores': temporal_scores,
        'ela_scores': ela_scores,
        'combined_scores': combined_scores
    }

# Enhanced function to analyze a video
def analyze_video(model, video_path, temp_folder="/content/temp_frames", batch_size=16):
    """
    Analyze a video for deepfake detection with enhanced background analysis.

    Args:
        model: Trained model
        video_path: Path to the video
        temp_folder: Folder to save temporary frames
        batch_size: Batch size for processing frames

    Returns:
        Dictionary with analysis results
    """
    # Extract frames
    frame_paths = extract_frames(video_path, temp_folder)

    if not frame_paths:
        raise ValueError("No frames could be extracted from the video")

    # Analyze frames in batches with enhanced analysis
    print("\nAnalyzing frames with enhanced background and temporal analysis...")
    analysis_results = batch_analyze_frames(model, frame_paths, batch_size)

    # Use the combined scores for final determination
    frame_predictions = analysis_results['combined_scores']

    # Calculate statistics
    avg_score = np.mean(frame_predictions)

    # Use 0.3 as the threshold, as requested
    threshold = 0.3

    binary_preds = [1 if p > threshold else 0 for p in frame_predictions]
    fake_count = sum(binary_preds)
    total = len(binary_preds)
    fake_percentage = (fake_count / total) * 100

    # Get confidence level
    confidence = abs(avg_score - 0.5) * 2  # Scale to 0-1

    # Plot distribution of different scores
    plt.figure(figsize=(15, 10))

    # Plot model predictions
    plt.subplot(2, 2, 1)
    plt.hist(analysis_results['model_predictions'], bins=10, alpha=0.7, color='blue')
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Model Prediction Scores')
    plt.xlabel('Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()

    # Plot background scores
    plt.subplot(2, 2, 2)
    plt.hist(analysis_results['background_scores'], bins=10, alpha=0.7, color='green')
    plt.title('Background Analysis Scores')
    plt.xlabel('Score (Higher = More Suspicious)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)

    # Plot temporal consistency scores
    plt.subplot(2, 2, 3)
    plt.hist(analysis_results['temporal_scores'], bins=10, alpha=0.7, color='orange')
    plt.title('Temporal Consistency Scores')
    plt.xlabel('Score (Higher = Less Consistent)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)

    # Plot combined scores
    plt.subplot(2, 2, 4)
    plt.hist(frame_predictions, bins=10, alpha=0.7, color='purple')
    plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Combined Scores (Final Decision)')
    plt.xlabel('Score (Higher = More Likely Fake)')
    plt.ylabel('Number of Frames')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.tight_layout()

    save_path = '/content/score_distributions.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/score_distributions.png'

    plt.savefig(save_path)
    plt.show()

    # Create second plot showing frame-by-frame scores
    plt.figure(figsize=(15, 8))
    frame_indices = range(len(frame_predictions))

    plt.plot(frame_indices, analysis_results['model_predictions'], 'b-', alpha=0.6, label='Model Score')
    plt.plot(frame_indices, analysis_results['background_scores'], 'g-', alpha=0.6, label='Background Score')
    plt.plot(frame_indices, analysis_results['temporal_scores'], 'y-', alpha=0.6, label='Temporal Score')
    plt.plot(frame_indices, frame_predictions, 'r-', linewidth=2, label='Combined Score')

    plt.axhline(y=threshold, color='purple', linestyle='--', label=f'Threshold ({threshold})')
    plt.title('Frame-by-Frame Analysis Scores')
    plt.xlabel('Frame Number')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_path2 = '/content/frame_analysis.png'
    if os.path.exists('/content/drive'):
        save_path2 = '/content/drive/MyDrive/frame_analysis.png'

    plt.savefig(save_path2)
    plt.show()

    return {
        'model_predictions': analysis_results['model_predictions'],
        'background_scores': analysis_results['background_scores'],
        'temporal_scores': analysis_results['temporal_scores'],
        'ela_scores': analysis_results['ela_scores'],
        'frame_predictions': frame_predictions,  # These are the combined scores
        'fake_count': fake_count,
        'total_frames': total,
        'fake_percentage': fake_percentage,
        'avg_score': avg_score,
        'confidence': confidence,
        'is_fake': fake_percentage > 50
    }

# Enhanced results printing function
def print_results(results):
    """Print the enhanced results of video analysis"""
    print("\n" + "="*60)



    print("🔍 ENHANCED DEEPFAKE DETECTION RESULTS 🔍")
    print("="*60)

    print(f"🧠 Detected {results['fake_count']}/{results['total_frames']} frames as FAKE")
    print(f"📊 Fake percentage: {results['fake_percentage']:.2f}%")
    print(f"⚖️ Average prediction score: {results['avg_score']:.4f}")
    print(f"🎯 Confidence level: {results['confidence']:.2f}")

    # Calculate average scores for each component
    avg_model = np.mean(results['model_predictions'])
    avg_bg = np.mean(results['background_scores'])
    avg_temp = np.mean(results['temporal_scores'])
    avg_ela = np.mean(results['ela_scores'])

    print("\n📈 DETECTION COMPONENT SCORES:")
    print(f"   🔹 Base model score: {avg_model:.4f}")
    print(f"   🔹 Background analysis: {avg_bg:.4f}")
    print(f"   🔹 Temporal consistency: {avg_temp:.4f}")
    print(f"   🔹 Error Level Analysis: {avg_ela:.4f}")

    # Identify which components contributed most to the detection
    scores = [
        ("Base model", avg_model),
        ("Background analysis", avg_bg),
        ("Temporal consistency", avg_temp),
        ("Error level analysis", avg_ela)
    ]

    # Sort by highest score (most suspicious)
    scores.sort(key=lambda x: x[1], reverse=True)

    print("\n🔎 TOP DETECTION SIGNALS:")
    for i, (name, score) in enumerate(scores[:2]):
        if score > 0.4:  # Only show significant signals
            print(f"   {i+1}. {name}: {score:.4f}")

    if results['is_fake']:
        print("\n🔴 VERDICT: This video is classified as FAKE")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")
    else:
        print("\n🟢 VERDICT: This video is classified as REAL")
        if results['confidence'] < 0.3:
            print("   ⚠️ Note: Low confidence detection")
        elif results['confidence'] > 0.7:
            print("   ✅ High confidence detection")

    print("="*60)
    print("📈 Check the saved distribution charts for more details")
    print("="*60)

# Enhanced function to visualize sample frames with their predictions
def plot_sample_frames(frames_dir, results, num_samples=5):
    """
    Plot sample frames from the video with their prediction scores and analysis

    Args:
        frames_dir: Directory containing video frames
        results: Dictionary with analysis results
        num_samples: Number of sample frames to show
    """
    import cv2
    import numpy as np

    # Get all frame files
    frame_files = sorted([f for f in os.listdir(frames_dir) if f.endswith('.jpg')])

    if len(frame_files) == 0:
        print("No frames found to display")
        return

    # Select sample frames (evenly distributed)
    if len(frame_files) <= num_samples:
        sample_indices = list(range(len(frame_files)))
    else:
        sample_indices = np.linspace(0, len(frame_files)-1, num_samples, dtype=int)

    plt.figure(figsize=(18, 4*num_samples))

    for i, idx in enumerate(sample_indices):
        frame_path = os.path.join(frames_dir, frame_files[idx])

        # Get all scores for this frame
        combined_score = results['frame_predictions'][idx]
        model_score = results['model_predictions'][idx]
        bg_score = results['background_scores'][idx]
        temp_score = results['temporal_scores'][idx] if idx < len(results['temporal_scores']) else 0
        ela_score = results['ela_scores'][idx]

        # Load and display the frame
        img = cv2.imread(frame_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

        # Perform ELA on the frame for visualization
        ela_img = error_level_analysis(frame_path)
        ela_img = cv2.cvtColor(ela_img, cv2.COLOR_BGR2RGB)

        # Detect faces to visualize segmentation
        faces = detect_faces(img)
        img_with_faces = img.copy()

        # Draw rectangles around faces
        for (x, y, w, h) in faces:
            cv2.rectangle(img_with_faces, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Get background mask
        bg_mask, background = segment_background(img, faces)

        # Create a subplot for this frame with 4 images
        plt.subplot(num_samples, 4, i*4+1)
        plt.imshow(img_with_faces)
        plt.title(f"Frame {idx}\nScore: {combined_score:.3f}",
                 color='red' if combined_score > 0.3 else 'green',
                 fontweight='bold')
        plt.axis('off')

        plt.subplot(num_samples, 4, i*4+2)
        plt.imshow(background)
        plt.title(f"Background\nScore: {bg_score:.3f}",
                 color='red' if bg_score > 0.5 else 'green')
        plt.axis('off')

        plt.subplot(num_samples, 4, i*4+3)
        plt.imshow(ela_img)
        plt.title(f"Error Level Analysis\nScore: {ela_score:.3f}",
                 color='red' if ela_score > 0.5 else 'green')
        plt.axis('off')

        # Plot a mini-dashboard of scores
        plt.subplot(num_samples, 4, i*4+4)
        score_labels = ['Combined', 'Model', 'Background', 'Temporal', 'ELA']
        score_values = [combined_score, model_score, bg_score, temp_score, ela_score]
        colors = ['red' if s > 0.3 else 'green' for s in score_values]

        y_pos = np.arange(len(score_labels))
        plt.barh(y_pos, score_values, color=colors)
        plt.yticks(y_pos, score_labels)
        plt.xlim(0, 1)
        plt.title("Component Scores")
        plt.grid(axis='x', alpha=0.3)

    plt.tight_layout()

    save_path = '/content/enhanced_sample_frames_analysis.png'
    if os.path.exists('/content/drive'):
        save_path = '/content/drive/MyDrive/enhanced_sample_frames_analysis.png'

    plt.savefig(save_path)
    plt.show()

    print("📊 Enhanced sample frames analysis saved")

# Enhanced function to save a detailed report
def save_detailed_report(video_name, results, elapsed_time):
    """
    Save a detailed analysis report to Google Drive

    Args:
        video_name: Name of the analyzed video
        results: Dictionary containing analysis results
        elapsed_time: Time taken for analysis (in seconds)
    """
    report_path = "/content/drive/MyDrive/deepfake_analysis_reports"
    os.makedirs(report_path, exist_ok=True)

    # Create timestamp
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    report_file = os.path.join(report_path, f"enhanced_analysis_{timestamp}_{video_name}.txt")

    with open(report_file, 'w') as f:
        f.write("=" * 70 + "\n")
        f.write(f"ENHANCED DEEPFAKE DETECTION REPORT - {timestamp}\n")
        f.write("=" * 70 + "\n\n")

        f.write(f"Video analyzed: {video_name}\n")
        f.write(f"Analysis date: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Analysis duration: {elapsed_time:.2f} seconds\n\n")

        f.write("RESULTS SUMMARY:\n")
        f.write("-" * 30 + "\n")
        f.write(f"Total frames analyzed: {results['total_frames']}\n")
        f.write(f"Frames detected as fake: {results['fake_count']}\n")
        f.write(f"Fake percentage: {results['fake_percentage']:.2f}%\n")
        f.write(f"Average prediction score: {results['avg_score']:.4f}\n")
        f.write(f"Confidence level: {results['confidence']:.2f}\n\n")

        # Add component analysis
        f.write("COMPONENT ANALYSIS:\n")
        f.write("-" * 30 + "\n")

        avg_model = np.mean(results['model_predictions'])
        avg_bg = np.mean(results['background_scores'])
        avg_temp = np.mean(results['temporal_scores'])
        avg_ela = np.mean(results['ela_scores'])

        f.write(f"Base model average score: {avg_model:.4f}\n")
        f.write(f"Background analysis average score: {avg_bg:.4f}\n")
        f.write(f"Temporal consistency average score: {avg_temp:.4f}\n")
        f.write(f"Error Level Analysis average score: {avg_ela:.4f}\n\n")

        if results['is_fake']:
            f.write("VERDICT: FAKE\n")
        else:
            f.write("VERDICT: REAL\n")

        if results['confidence'] < 0.3:
            f.write("Note: Low confidence detection\n")
        elif results['confidence'] > 0.7:
            f.write("High confidence detection\n")

        f.write("\nDETAILED FRAME ANALYSIS:\n")
        f.write("-" * 30 + "\n")

        # Update frame-level statistics using 0.3 threshold
        frames_fake = sum(1 for p in results['frame_predictions'] if p > 0.3)
        frames_real = sum(1 for p in results['frame_predictions'] if p <= 0.3)

        f.write(f"FAKE frames (score > 0.3): {frames_fake} frames\n")
        f.write(f"REAL frames (score <= 0.3): {frames_real} frames\n\n")

        f.write("First 20 frame predictions with component breakdown:\n")
        f.write("Frame\tCombined\tModel\tBackground\tTemporal\tELA\tVerdict\n")

        # Write the first 20 frames details
        num_frames_to_show = min(20, len(results['frame_predictions']))
        for i in range(num_frames_to_show):
            combined = results['frame_predictions'][i]
            model = results['model_predictions'][i]
            bg = results['background_scores'][i] if i < len(results['background_scores']) else 0
            temp = results['temporal_scores'][i] if i < len(results['temporal_scores']) else 0
            ela = results['ela_scores'][i] if i < len(results['ela_scores']) else 0

            verdict = "FAKE" if combined > 0.3 else "REAL"

            f.write(f"{i}\t{combined:.3f}\t{model:.3f}\t{bg:.3f}\t{temp:.3f}\t{ela:.3f}\t{verdict}\n")

        f.write("\nANALYSIS INTERPRETATION:\n")
        f.write("-" * 30 + "\n")

        # Interpretation based on component scores
        if avg_model > 0.5:
            f.write("- The base deepfake detection model strongly indicates manipulation\n")

        if avg_bg > 0.5:
            f.write("- Background inconsistencies detected, suggesting possible compositing\n")

        if avg_temp > 0.5:
            f.write("- Temporal inconsistencies detected between frames\n")

        if avg_ela > 0.5:
            f.write("- Error Level Analysis shows compression artifacts consistent with manipulation\n")

        # Overall assessment
        f.write("\nOVERALL ASSESSMENT:\n")
        f.write("-" * 30 + "\n")

        if results['is_fake']:
            if results['confidence'] > 0.7:
                f.write("This video shows strong evidence of being manipulated or generated.\n")
            elif results['confidence'] > 0.3:
                f.write("This video shows moderate evidence of manipulation, but further analysis may be needed.\n")
            else:
                f.write("This video shows weak indicators of manipulation, but results are inconclusive.\n")
        else:
            if results['confidence'] > 0.7:
                f.write("This video shows strong indicators of being authentic.\n")
            elif results['confidence'] > 0.3:
                f.write("This video appears authentic, but with moderate confidence.\n")
            else:
                f.write("This video lacks strong manipulation indicators, but results are inconclusive.\n")

        f.write("\n" + "=" * 70 + "\n")
        f.write("End of report\n")
        f.write("=" * 70)

    print(f"📝 Detailed report saved to {report_file}")
if __name__ == "__main__":
   detect_deepfake()
```
